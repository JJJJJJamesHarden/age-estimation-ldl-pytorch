```python
from PIL import Image
```

```python
def show_img(info, ratio = 0.3):
    info = info.split(",")
    img = Image.open(info[0])
    img = img.resize((int(img.size[0]*ratio), int(img.size[1]*ratio)))
    display(img)
    print("Real Age:", info[1], "\t Pred Age:", info[2], "\tAbsolute Error:", info[3])
```
